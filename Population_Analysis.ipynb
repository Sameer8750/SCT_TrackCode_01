**Dashboard**

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create subplots
fig = make_subplots(rows=3, cols=2,
                    specs=[[{"type": "bar"}, {"type": "histogram"}],
                           [{"type": "scatter"}, {"type": "box"}],
                           [{"type": "pie"}, None]],
                    subplot_titles=("Rank Distribution", "2022 Population Distribution",
                                    "Population vs Rank", "Population by Continent",
                                    "World Population Percentage"))

# Bar chart for Rank Distribution
rank_counts = df['Rank'].value_counts()
fig.add_trace(go.Bar(x=rank_counts.index, y=rank_counts.values), row=1, col=1)

# Histogram for 2022 Population Distribution
fig.add_trace(go.Histogram(x=df['2022 Population']), row=1, col=2)

# Scatter plot for Population vs Rank
fig.add_trace(go.Scatter(x=df['Rank'], y=df['2022 Population'], mode='markers'), row=2, col=1)

# Box plot for Population by Continent
fig.add_trace(go.Box(x=df['Continent'], y=df['2022 Population']), row=2, col=2)

# Pie chart for World Population Percentage
top_countries = df.nlargest(10, 'World Population Percentage')
others = pd.DataFrame([['Others', top_countries['World Population Percentage'].sum()]], columns=['Country/Territory', 'World Population Percentage'])
top_countries = pd.concat([top_countries, others], ignore_index=True)
fig.add_trace(go.Pie(labels=top_countries['Country/Territory'], values=top_countries['World Population Percentage']), row=3, col=1)

# Update layout
fig.update_layout(height=1000, width=1200, title_text="World Population Dashboard")
fig.show()
```
